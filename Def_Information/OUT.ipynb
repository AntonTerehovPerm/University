2. Ищем самый часто встречающийся символ - он будет 'пробел'

```python
def zap(spis_b, x):#запись всех символов и количества их встреч в словарь
    #print(x)
    if spis_b.get(x) == None:
        spis_b[x] = 1
    else:
        spis_b[x] += 1
    #return spis_b
```

```python
def chistot_symbols(input_):#парсим строку и складываем ее в словарь
    import operator
    spis_b = {}
    [zap(spis_b,x) for x in [input_[i - n:i] for i in range(n, len(input_) + 1, n)]]
    return sorted(spis_b.items(), key=operator.itemgetter(1), reverse = True)#сортируем словарь по значениям
```

<b>2. Определение длины ключа<br></b>

Мы знаем, что при XOR двух одинаковых чисел, получается 0. Если попробовать XOR часть текста определенной длины на сам текст, как в повторяющемся ключе, то можно по наибольшему количеству 0 или проще говоря переходу за грань определить длину ключа.<br>

```python
def checked_leng_key(text):
    max_symb = [0,0]#максимально значение % и номер сдвига
    for i in range(1, int(100)+1):#берем длину ключа 100, потому что люди не будут делать ключ длиней, чем 100 символов
        text_obrez = text[-i:]+text[:-i]
        step = convert.shifr_duplicate_key(text, text_obrez)
        percent = round(100*convert.chistot_symbols(step)[0][1]/len(text),4)
        if percent > max_symb[0]:
            max_symb[0] = percent
            max_symb[1] = i
        #print(f'Смещение: {i}, Процент: {percent}')
    print(f'Максимальный процент повторяющихся символов: {max_symb[0]}, Найден при смещении: {max_symb[1]}')
    return max_symb[1]
```

```python
key_leng = checked_leng_key(byte_text)
```

<b>3. Определяем символы ключа и дешифруем текст<br></b>

3.1 разбиваем текст на подпоследовательности длины ключа, по каждому символу( 1 символ из каждой строки массива всех строк - новая строка, второй символ из каждой строки массива всех строк - новая строка)

```python
def chasti_text(text,key):
    massiv_chastei = ['' for x in range(key)]            
    for x in range(0, len(text), key):
        symbol = text[x:x+key]
        for y in range(len(symbol)):
            massiv_chastei[y] += symbol[y]
    return massiv_chastei 
```
